```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'https://github.com/kaopanboonyuen/Python-Data-Science/raw/master/Dataset/adult_data/adult.csv')
df.head(30)
```

### Data Cleansing

```python
# แทนที่ "?" ในคอลัมน์ workclass ด้วยค่าโหมด
mode_workclass = df['workclass'].mode().iloc[0]  # หาโหมดของคอลัมน์ workclass
df['workclass'].replace("?", mode_workclass, inplace=True)  # แทนที่ "?" ด้วยโหมด

# แทนที่ "?" ในคอลัมน์ occupation ด้วยค่าโหมด
mode_occupation = df['occupation'].mode().iloc[0]  # หาโหมดของคอลัมน์ occupation
df['occupation'].replace("?", mode_occupation, inplace=True)  # แทนที่ "?" ด้วยโหมด

# แทนที่ "?" ในคอลัมน์ native-country ด้วยค่าโหมด
mode_nativecountry = df['native-country'].mode().iloc[0]  # หาโหมดของคอลัมน์ native-country
df['native-country'].replace("?", mode_nativecountry, inplace=True)  # แทนที่ "?" ด้วยโหมด

```

```python
print('mode_workclass =', mode_workclass)
print('mode_occupation =', mode_occupation)
print('mode_nativecountry =', mode_nativecountry)
```

```python
df.head(30)
```

```python
df.fnlwgt = df.fnlwgt.astype('string')
df.info()
```

## Converting Categorical Features

We'll need to convert categorical features to dummy variables using pandas! Otherwise our machine learning algorithm won't be able to directly take in those features as inputs.

```python
Workclass = pd.get_dummies(df['workclass'],drop_first=True)
Education = pd.get_dummies(df['education'],drop_first=True)
Maritalstatus = pd.get_dummies(df['marital-status'],drop_first=True)
Occupation = pd.get_dummies(df['occupation'],drop_first=True)
Relationship = pd.get_dummies(df['relationship'],drop_first=True)
Race = pd.get_dummies(df['race'],drop_first=True)
Gender = pd.get_dummies(df['gender'],drop_first=True)
Nativecountry = pd.get_dummies(df['native-country'],drop_first=True)
Income  = pd.get_dummies(df['income'],drop_first=True)

df.head()
```

```python
df.drop(['workclass'],axis=1,inplace=True)
df.drop(['education'],axis=1,inplace=True)
df.drop(['marital-status'],axis=1,inplace=True)
df.drop(['occupation'],axis=1,inplace=True)
df.drop(['relationship'],axis=1,inplace=True)
df.drop(['race'],axis=1,inplace=True)
df.drop(['gender'],axis=1,inplace=True)
df.drop(['native-country'],axis=1,inplace=True)
df.drop(['income'],axis=1,inplace=True)

df.drop(['fnlwgt'],axis=1,inplace=True)
df.head()
```

```python
df = pd.concat([df,Workclass],axis=1)
df = pd.concat([df,Education],axis=1)
df = pd.concat([df,Maritalstatus],axis=1)
df = pd.concat([df,Occupation],axis=1)
df = pd.concat([df,Relationship],axis=1)
df = pd.concat([df,Race],axis=1)
df = pd.concat([df,Gender],axis=1)
df = pd.concat([df,Nativecountry],axis=1)
df = pd.concat([df,Income],axis=1)

df.head()
```

```python
df.info()
```

### Building a Logistic Regression model

```python
from sklearn.model_selection import train_test_split
X = df.drop('>50K',axis=1)
y = df['>50K']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.30, random_state=101)
```

### Training and Predicting

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(list(y_test[:5]))
print(y_pred[:5])
```

## Evaluation

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
confusion_matrix(y_test, y_pred, labels=[0,1])
```

```python
print(classification_report(y_test,y_pred, digits=4))
```
