```python
import pandas as pd
df = pd.read_csv('Train.csv')
df.dropna(inplace=True)
```

```python
df['class'].value_counts()
```

```python
pos_data=df['message']
pos_target=df['class']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB

vectorizer=CountVectorizer()
tfidf=TfidfTransformer()
X=vectorizer.fit_transform(pos_data)
X=tfidf.fit_transform(X)
clf=MultinomialNB().fit(X,pos_target)
```

```python
docs_new = ['สุกรี เขียน โปรแกรม python','สุกรี พบ ลุง โกง','สุกรี ดู youtube','สุกรี เลี้ยง แมว','สุกรี นอน บน เตียง',
            'สุกรี ถ่าย รูป ทะเล','สุกรี ยืม นาฬิกา เพื่อน','สุกรี ตรวจ การบ้าน','สุกรี เป็น นายก','สุกรี ออก รายการ super10' ]
X_new = vectorizer.transform(docs_new)
predicted = clf.predict(X_new)
X_new = tfidf.transform(X_new)
print(predicted)
```
